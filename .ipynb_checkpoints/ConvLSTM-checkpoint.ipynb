```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import ConvLSTM2D, Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('smart_traffic_management_dataset.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')

label_encoders = {}
for col in ['weather_condition', 'signal_status']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

features = df.drop(columns=['timestamp', 'traffic_volume'])
target = df['traffic_volume'].values.reshape(-1, 1)

scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(features)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(target)

```

```python
def create_sequences(data, seq_len):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len, :-1])
        y.append(data[i+seq_len, -1])
    return np.array(X), np.array(y)
```

```python
data_scaled = np.hstack((X_scaled, y_scaled))
SEQ_LEN = 10
X_seq, y_seq = create_sequences(data_scaled, SEQ_LEN)
```

```python
X_conv = X_seq.reshape((X_seq.shape[0], SEQ_LEN, 10, 1, 1))
X_train, X_test, y_train, y_test = train_test_split(X_conv, y_seq, test_size=0.2, random_state=42)

model_conv = Sequential([
    ConvLSTM2D(filters=64, kernel_size=(3, 1), activation='relu', input_shape=(SEQ_LEN, 10, 1, 1)),
    Flatten(),
    Dense(50, activation='relu'),
    Dense(1, activation='linear')
])
model_conv.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_conv = model_conv.fit(X_train, y_train, epochs=20, batch_size=32,
                              validation_data=(X_test, y_test),
                              callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
                              verbose=1)
```

```python
y_pred = model_conv.predict(X_test)
y_pred_inv = scaler_y.inverse_transform(y_pred)
y_test_inv = scaler_y.inverse_transform(y_test.reshape(-1, 1))
```

```python
plt.plot(y_test_inv[:100], label='Actual')
plt.plot(y_pred_inv[:100], label='Predicted')
plt.title("ConvLSTM Prediction")
plt.legend()
plt.show()
```
